```python
# Genera la tabla crosstab
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])

# Asegurarnos de que ambos valores True y False existen en todas las categorÃ­as para evitar errores
crosstab_result[True] = crosstab_result.get(True, 0)
crosstab_result[False] = crosstab_result.get(False, 0)

# Calcular el ratio True/False para cada categorÃ­a
crosstab_result['ratio'] = crosstab_result[True] / crosstab_result[False]

# Si queremos manejar casos donde False sea 0, podemos usar fillna o hacer algo diferente
crosstab_result['ratio'] = crosstab_result['ratio'].replace([float('inf'), float('nan')], -1)

crosstab_sorted = crosstab_result.sort_values(by='ratio', ascending=False)


# Mostrar los resultados con el ratio
print(crosstab_sorted.head(10))
```

```python
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])

# Asegurarnos de que ambas columnas True y False existan
crosstab_result[True] = crosstab_result.get(True, 0)
crosstab_result[False] = crosstab_result.get(False, 0)

# Calcular el ratio True/False para cada categorÃ­a
crosstab_result['ratio'] = crosstab_result[True] / crosstab_result[False]

# Si False es cero, manejar esos casos
crosstab_result['ratio'].replace([float('inf'), float('nan')], -1, inplace=True)

# Ordenar por el ratio en orden descendente
crosstab_sorted = crosstab_result.sort_values(by='ratio', ascending=False)

# Seleccionar las primeras 10 categorÃ­as
top_10 = crosstab_sorted.head(10)

# Graficar solo estas categorÃ­as
top_10.drop(columns='ratio').plot(kind="bar", stacked=True)

plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Top 10 Categories with Highest True/False Ratio')
plt.show()
```

```python
#Create bins to be able to group publications according to surface area
bins = np.arange(0, df['price'].max() + 10000, 10000)
bins
```

```python
df['price_range'] = pd.cut(df['price'], bins=bins)

count_ranges = df['price_range'].value_counts().sort_index()

print(count_ranges)
```

```python
#Identify ranges of outliers according to prices
values = df['price'].dropna()

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = values.quantile(0.25)
Q3 = values.quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Determine the outlier limits
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print(upper_bound)

# Identify the outliers
outliers = values[(values < lower_bound) | (values > upper_bound)]

print("Outliers in 'price':")
print(outliers)
```

```python
#Eliminate outliers according to housing surface
df = df[df['price'] <= upper_bound]
```

```python
# Define the bins using numpy for easier calculation
bins = np.arange(0, df['price'].max() + 2, 2)

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=bins, color='skyblue', edgecolor='black')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Number of Listings')
plt.xticks(bins, rotation=45)  # To improve readability of x-axis labels
plt.show()
```

```python
# Frequency table for 'category'
cat_frequency_table = df['category'].value_counts().head()

# Calculating the proportion of each unique value in the 'category'
cat_proportion_table = df['category'].value_counts(normalize=True).head() # Muestra la frecuencia relativa (%)

cat_frequency_table, cat_proportion_table
```
